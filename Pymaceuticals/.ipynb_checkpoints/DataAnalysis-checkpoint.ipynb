### Analysis
* Overall, it is clear that Capomulin and Ramicane outperform all other treatment options in the screen.
* Capomulin and Ramicane were the only treatments to reduce tumor volume. It held to a 19% reduction in tumor volume over the course of trial, whereas all other drugs were correlated with an increase in tumor volume by roughly 40-50%.
* Capomulin greatly limited the spread of the tumor compared to other treatment options. By study end, the average mouse on Capomulin had only 1 new metastatic site, as opposed to the average 2-3 found in mice of other treatment options.
* Lastly, mice on the Capomulin treatment had the highest survival rate of any treatment in the screen. Over 90% of mice treated by Capomulin survived the full duration of the trial, compared to only 35-45% of mice on other treatment options. 

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Read Data
mouse_drug_data = pd.read_csv("data/mouse_drug_data.csv")
clinical_data = pd.read_csv("data/clinicaltrial_data.csv")
# Combine the data into a single dataset
clinical_data_complete = pd.merge(clinical_data, mouse_drug_data, how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview
clinical_data_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_vols_mean = clinical_data_complete.groupby(['Drug','Timepoint']).mean()['Tumor Volume (mm3)']
tumor_vols_mean = pd.DataFrame(tumor_vols_mean).reset_index()
tumor_vols_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vols_sem = clinical_data_complete.groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)']
tumor_vols_sem = pd.DataFrame(tumor_vols_sem).reset_index()
tumor_vols_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_vols_pivot_mean = tumor_vols_mean.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
tumor_vols_pivot_sem = tumor_vols_sem.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
```

```python
tumor_vols_pivot_mean.head(3)
```

```python
tumor_vols_pivot_sem.head(3)
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Capomulin"], yerr=tumor_vols_pivot_sem["Capomulin"], color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Ceftamin"], yerr=tumor_vols_pivot_sem["Ceftamin"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Infubinol"], yerr=tumor_vols_pivot_sem["Infubinol"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Ketapril"], yerr=tumor_vols_pivot_sem["Ketapril"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Naftisol"], yerr=tumor_vols_pivot_sem["Naftisol"], color="c", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Placebo"], yerr=tumor_vols_pivot_sem["Placebo"], color="m", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Propriva"], yerr=tumor_vols_pivot_sem["Propriva"], color="y", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Ramicane"], yerr=tumor_vols_pivot_sem["Ramicane"], color="r", marker="s", markersize=5, linestyle="dashed", linewidth=0.50,alpha=0.5)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Stelasyn"], yerr=tumor_vols_pivot_sem["Stelasyn"], color="b", marker="d", markersize=5, linestyle="dashed", linewidth=0.50,alpha=0.5)
plt.errorbar(tumor_vols_pivot_mean.index, tumor_vols_pivot_mean["Zoniferol"], yerr=tumor_vols_pivot_sem["Zoniferol"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50,alpha=0.5)

plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.ylim(30,85)
plt.xlabel("Time (Days)")
plt.grid(axis='y')
plt.legend(['Capomulin','Ceftamin','Infubinol','Ketapril','Naftisol','Placebo','Propriva','Ramicane','Stelasyn','Zoniferol'],loc="upper left", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig("output/Fig1.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_sites_mean = clinical_data_complete.groupby(['Drug','Timepoint']).mean()['Metastatic Sites']
met_sites_mean = pd.DataFrame(met_sites_mean)
met_sites_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sites_sem = clinical_data_complete.groupby(['Drug','Timepoint']).sem()['Metastatic Sites']
met_sites_sem = pd.DataFrame(met_sites_sem)
met_sites_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_sites_pivot_mean = met_sites_mean.reset_index()
met_sites_pivot_sem = met_sites_sem.reset_index()
met_sites_pivot_mean = met_sites_pivot_mean.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]
met_sites_pivot_sem = met_sites_pivot_sem.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]
```

```python
met_sites_pivot_sem.head(3)
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Capomulin"], yerr=met_sites_pivot_sem["Capomulin"], color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Ceftamin"], yerr=met_sites_pivot_sem["Ceftamin"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Infubinol"], yerr=met_sites_pivot_sem["Infubinol"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Ketapril"], yerr=met_sites_pivot_sem["Ketapril"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Naftisol"], yerr=met_sites_pivot_sem["Naftisol"], color="c", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Placebo"], yerr=met_sites_pivot_sem["Placebo"], color="m", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Propriva"], yerr=met_sites_pivot_sem["Propriva"], color="y", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Ramicane"], yerr=met_sites_pivot_sem["Ramicane"], color="r", marker="s", markersize=5, linestyle="dashed", linewidth=0.50,alpha=0.5)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Stelasyn"], yerr=met_sites_pivot_sem["Stelasyn"], color="b", marker="d", markersize=5, linestyle="dashed", linewidth=0.50,alpha=0.5)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Zoniferol"], yerr=met_sites_pivot_sem["Zoniferol"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50,alpha=0.5)

plt.title("Metastatic Spread During Treatment")
plt.ylabel("Met. Sites")
plt.xlabel("Treatment Duration (Days)")
plt.grid()
plt.legend(['Capomulin','Ceftamin','Infubinol','Ketapril','Naftisol','Placebo','Propriva','Ramicane','Stelasyn','Zoniferol'],loc="upper left", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig("output/Fig2.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_count = clinical_data_complete.groupby(['Drug','Timepoint']).count()['Mouse ID']
survival_count = pd.DataFrame({"Mouse Count": survival_count})
#clinical_data_complete
survival_count.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_count_pivot = survival_count.reset_index()
survival_count_pivot = survival_count_pivot.pivot(index='Timepoint',columns='Drug')['Mouse Count']
survival_count_pivot.head(3)
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(survival_count_pivot['Capomulin']/survival_count_pivot['Capomulin'][0]*100, "ro", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Ceftamin']/survival_count_pivot['Ceftamin'][0]*100, "b^", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Infubinol']/survival_count_pivot['Infubinol'][0]*100, "gs", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Ketapril']/survival_count_pivot['Ketapril'][0]*100, "kd", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Naftisol']/survival_count_pivot['Naftisol'][0]*100, "cX", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Placebo']/survival_count_pivot['Placebo'][0]*100, "m*", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Propriva']/survival_count_pivot['Propriva'][0]*100, "yH", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Ramicane']/survival_count_pivot['Ramicane'][0]*100, "rP", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Stelasyn']/survival_count_pivot['Stelasyn'][0]*100, "bH", linestyle="dashed", markersize=5,linewidth=0.50)
plt.plot(survival_count_pivot['Zoniferol']/survival_count_pivot['Zoniferol'][0]*100, "gH", linestyle="dashed", markersize=5,linewidth=0.50)

plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid()
plt.legend(['Capomulin','Ceftamin','Infubinol','Ketapril','Naftisol','Placebo','Propriva','Ramicane','Stelasyn','Zoniferol'],loc="best", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig("output/Fig3.png")
# Show the Figure
plt.show()
```
